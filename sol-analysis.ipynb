```python
plt.figure(figsize=(15, 10))
f, ax = plt.subplots(1)
ax.set_ylim(ymin=-0.125)
ax.set_ylim(ymax=0.125)
plt.plot(opalstat['s'],100*opalstat['rms_x'],label='x *100 OPAL')
plt.plot(opalstat['s'],100*opalstat['rms_y'],label='y *100 OPAL')
plt.plot(opalstat['s'],100*opalstat['rms_s'],label='z *100 OPAL')
plt.plot(opalstat['s'],opalstat['Bz_ref'],label='Bz OPAL')
plt.xlabel('s [m]')
plt.ylabel('rms x [m]')
plt.legend()
plt.show()
```
